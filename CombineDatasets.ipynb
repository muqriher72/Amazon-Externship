```python
import pandas as pd

glassdoor_df = pd.read_csv('glassdoor_sentiment.csv')
youtube_df = pd.read_csv('youtube_sentiment.csv')
```

```python
glassdoor_df.rename(columns={'cleaned_summary': 'cleaned_text'}, inplace=True)
youtube_df.rename(columns={'cleaned_transcript': 'cleaned_text'}, inplace=True)
```

```python
glassdoor_df['platform'] = 'Glassdoor'
youtube_df['platform'] = 'YouTube'
```

```python
combined_df = pd.concat([glassdoor_df, youtube_df], ignore_index=True)
```
