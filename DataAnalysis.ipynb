```python
# Importing pandas
import pandas as pd
```

```python
# Import filles
from google.colab import files
uploaded = files.upload()

df = pd.read_csv("glassdoor_reviews.csv")
```

```python
# Inspect the first few rows of the DataFrame
print("First five rows of the dataset:")
print(df.head())

# Get a summary of the DataFrame
print("\nDataFrame summary:")
print(df.info())
```

```python
# Explore the dataset for duplicates, missing, and non-unique values

df['company_name'].value_counts()

# Check how many values are missing in each column
df.isnull().sum().sort_values(ascending=False)
```

```python
# Check how many values are unique in each column
df.nunique().sort_values()
```

```python
# Since all reviews are for Amazon Fulfillment centers, drop that column along with review advice and career opportunities rating
df = df.drop(columns=['company_name', 'review_advice', 'career_opportunities_rating'])

# Check new dataset
df.head()
```

```python
# Drop rows where the main review text is missing
df = df.dropna(subset=['summary'])
```

```python
# Remove any exact duplicate rows
df = df.drop_duplicates()
```

```python
df.info()
```

```python
# Save the cleaned DataFrame to a new CSV
df.to_csv('glassdoor_cleaned.csv', index=False)
```
